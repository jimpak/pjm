```python
from folium import plugins
import os
N = 100
data = np.array(
    [np.random.uniform(low=35,high=38, size=N),
     np.random.uniform(low=126, high=129, size=N)]
).T
popup=[str(i+1) for i in range(N)]
m=folium.Map(location=[lat1, lng1], zoom_start=6)
plugins.MarkerCluster(data, popups=popup).add_to(m)
m.save(os.path.join('data', 'plugins_1.html')) # data/plugins_1.html
m
```

```python
df = pd.read_csv('data/CCTV_20190917.csv')
df
```

```python
df1=df.loc[df['카메라대수']>=1, :] # 카메라대수 0 제거
df1
```

```python
popup = list(df1['소재지도로명주소'])

# 모든 데이터의 중간값 구하기.
lat_avg=np.array(list(df1['위도'])).mean()
lng_avg=np.array(list(df1['경도'])).mean()

data=np.array([df1['위도'], df1['경도']]).T
print(popup)
print(data)
print(lat_avg)
print(lng_avg)
```

```python
m=folium.Map(location=[lat_avg, lng_avg], zoom_start=11)
plugins.MarkerCluster(data, popups=popup).add_to(m)
m.save(os.path.join('data','cctv.html'))
m
```
