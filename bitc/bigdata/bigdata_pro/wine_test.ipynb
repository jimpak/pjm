# 회귀분석 
- 회귀분석 식 : y= a1x1+a2x2+a3x3....+b

```python
formula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
      residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + \
      density + pH + sulphates + alcohol'
```

```python
regression_result=ols(formula, data=wine).fit() # 훈련모델
```

```python
regression_result.summary()
```

```python
sample=wine[wine.columns.difference(['quality','type'])]
sample1=sample[0:5][:]
sample1
```

```python
sample_predict=regression_result.predict(sample1)
```

```python
print(sample_predict)
```
