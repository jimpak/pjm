```python
from konlpy.tag import *
hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
okt = Okt()
```

```python
reviews = open('data/yes24_review.txt', 'r').readlines() # 블럭단위로 리스트 출력.
review_str=''
for review in reviews:
    review_str += review + ' ' # 하나의 문단으로 변경
review_str
```

```python
nouns = okt.nouns(review_str) # 명사만 추출
nouns1=[]
for noun in nouns:
    if len(noun) > 1: # 글자가 2글자 이상인 단어만 출력
        nouns1.append(noun)
nouns1
```

```python
from collections import Counter
c1 = Counter(nouns1)
c1
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
font_path='data/fonts/MALGUNBD.TTF'
wc=WordCloud(width=1000, height=1000, background_color='white', font_path=font_path)
plt.imshow(wc.generate_from_frequencies(c1))
plt.axis('off')
plt.savefig('data/images/wc.png')
plt.show()
```
