```python
import numpy as np
import pandas as pd
```

```python
data_url='http://lib.stat.cmu.edu/datasets/boston'
raw_df=pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
raw_df
```

```python
data=np.hstack([raw_df.values[::2], raw_df.values[1::2, :2]])
data
```

```python
target=raw_df.values[1::2, 2]
target
```

```python
df=pd.DataFrame(data,columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
        'TAX', 'PTRATIO', 'B', 'LSTAT'])
df
```

```python
df['TARGET']=target
df
```

요약통계
1. CHAS에 따른 집가격(TARGET)의 평균, 표준편차, 중앙값을 구하라.
2. CHAS에 따른 집가격의 평균차이로 T-test를 수행하여, 차이가 있는지 pvalue를 확인한다.
3. 선형회귀분석을 모델을 생성하고 모델을 사용하여 예측값을 구하라.
4. 새로운 데이터 2개를 사용하여 예측값을 구하라.
5. 평균 주택가격(TARGET)를 사용하여 histplot를 커널 밀도 추정(KDE)를 적용하여 작성한다.
6. 부분회귀분석 시각화를 작성하여 독립변수들이 종속변수에 미치는 영향을 분석하라. 
 - 1개독립변수 분석(plot_partregress)은 (CRIM)으로한다.

```python
from scipy import stats
chas_0=df.loc[df['CHAS']==0, 'TARGET']
chas_1=df.loc[df['CHAS']==1, 'TARGET']
print(chas_0.shape)
print(chas_1.shape)
```

```python
stats.ttest_ind(chas_0, chas_1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['TARGET','RM','AGE','CRIM','DIS','TAX']])
```

```python
from statsmodels.formula.api import ols
#y=a1x1+a2x2+....+b
formula='TARGET ~ CRIM + ZN + INDUS + RM + AGE + DIS + RAD + TAX + PTRATIO + LSTAT'
regression_result=ols(formula, data=df).fit()
regression_result.summary()
```

```python
sample=df[df.columns.difference(['TARGET','B','CHAS'])]
sample1=sample[10:20][:]
sample1
```

```python
sample1_predict=regression_result.predict(sample1)
sample1_predict
```

```python
df1=df.TARGET
df1[10:20]
```

```python
result=pd.concat([sample1_predict,df1[10:20]], axis=1)
result
```
