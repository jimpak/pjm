```python
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
font_path='data/fonts/MALGUN.TTF'
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
kolaw_nouns = Text(okt.nouns(c))
kolaw_nouns.plot(30)
```

```python
kv = kolaw_nouns.vocab()
kv
```

```python
kv2 = dict()
for tag, count in kv.items():
    if (len(tag))>1:
        kv2[tag]=count
kv2
```

```python
from wordcloud import WordCloud
font_path1 = 'data/fonts/MALGUNBD.TTF'
wc = WordCloud(width=800, height=800, background_color='white', font_path=font_path1, random_state=0, colormap='Accent_r')
plt.imshow(wc.generate_from_frequencies(kv2))
plt.axis('off')
```

```python
from collections import Counter
```

```python
s1 = "aabbccccddd"
list1 = ['abc', 'abc', 'aa', 'aa', 'bb', 'bb', 'bb']
counter1 = Counter(s1)
counter2 = Counter(list1)
print(counter1)
print(counter2)
```

```python
nouns1 = okt.nouns(c)
noun_count1 = Counter(nouns1)
noun_count2 = noun_count1.most_common(20) # 빈도수 제일 높은 토큰만 출력 (튜플로 결과 출력됨.)
print(noun_count1.items()) # noun_count2는 딕셔너리가 아니기 때문에 이 방식은 오류발생
print('counter2:', noun_count2)
count3 = dict()
for key, value in noun_count1.items():
    if(len(key) > 1 and value > 1):
        count3[key] = value
print('count3:', count3)
```

```python
wc = WordCloud(width=1000, height=1000, background_color='white', random_state=0, font_path=font_path1, colormap='jet')
plt.imshow(wc.generate_from_frequencies(count3))
plt.axis('off')
plt.savefig('data/images/wc00.png')
plt.show()
```
