```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import json
import urllib.request # requests를 사용해도 됨.

client_id = "ijm7fpfcH9iWm5LBifhk"
client_secret = "LUilHrShRi"
encText = urllib.parse.quote("빅데이터")

url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "display=100" # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과

# urllib.request.urlopen(url)
request = urllib.request.Request(url) # request 객체 생성
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

response = urllib.request.urlopen(request)
rescode = response.getcode()

str1 = ""
if(rescode==200): # response로 가져온 객체가 존재한다면
    response_body = response.read()
    # print(response_body)

    # print(response_body.decode('utf-8'))
    # items=response_body.decode('utf-8')['items'] # items를 읽어오지 못함.

    res_body = json.loads(response_body.decode('utf-8'))
    items = res_body['items'] # json으로 객체를 저장하면 items를 읽어올수 있음.
    for item in items:
        str1 += item['title']
        str1 += item['description']
        # print(item['title'])
        # print(item['description'])
        # print()
    print(str1)
else:
    print("Error Code:" + rescode)
```

```python
from konlpy.tag import *
okt = Okt()
nouns = okt.nouns(str1)
nouns
```

```python
from collections import Counter
c1 = Counter(nouns)
cd = dict()
for key, value in c1.items():
    # print(key, value)
    if(len(key) > 1 and value > 1):
        cd[key] = value

# c2 = c1.most_common(20)
# c2
```
