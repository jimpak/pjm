```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url = 'https://www.yes24.com/Product/Search?domain=ALL&query=python'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
book_list = soup.select("#yesSchList > li")
book_list1 = []
for book in book_list:
    title = book.select_one('div > div.item_info > div.info_row.info_name > a')
    if(title != None):
        title_str = title.text
        print(title.text)
    else:
        title_str = ''

    auth = book.select_one('div >  div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a')
    if(auth != None):
        auth_str = auth.text
        print(auth.text)
    else:
        auth_str = ''

    pub = book.select_one('div >  div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub')
    if(pub != None):
        pub_str = pub.text
        print(pub.text)
    else:
        pub_str = ''

    price = book.select_one('div > div.item_info > div.info_row.info_price > strong')
    if(price != None):
        price_str = price.text
        print(price.text)
    else:
        price_str = ''

    point = book.select_one('div > div.item.info > div.info_row.info_price > span.yPoint')
    if(point != None):
        point_str = point.text
        print(point.text)
    else:
        point_str = '0'
    
    book_list1.append([title_str, auth_str, pub_str, price_str, point_str])
    print('-------------')

book_list1
# book_list
# li_list = soup.find_all('li')
# li_list
```
