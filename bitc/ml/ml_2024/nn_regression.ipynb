```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
model=nn.Linear(1,1) # input_dim=1, output_dim=1  y=wx+b
print(list(model.parameters()))
```

```python
optimizer=optim.SGD(model.parameters(), lr=0.01)
epochs=2001

for epoch in range(epochs):
    h=model(x_data)
    cost=F.mse_loss(h,t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch%100==0:
        print('epoch:',epoch, 'cost:',cost.item())
```

## Class 사용

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self): # 모델생성
        super().__init__()
        self.linear=nn.Linear(1,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model = LinearRegressionModel()
print(list(model.parameters()))
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 2001
for epoch in range(epochs):
    y = model(x_data)
    cost = F.mse_loss(y, t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100 ==0:
        print(epoch, cost.item(), list(model.parameters()))
```

```python
class MultiLinearRegressionModel(nn.Module):
    def __init__(self): # 모델생성
        super().__init__()
        self.linear=nn.Linear(3,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model2 = MultiLinearRegressionModel()
print(list(model2.parameters()))
optimizer = optim.SGD(model2.parameters(), lr=0.00001)
```

```python
epochs = 200001
for epoch in range(epochs):
    y = model2(x_data3)
    cost = F.mse_loss(y, t_data3)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%400 ==0:
        print(epoch, cost.item(), list(model2.parameters()))
```
