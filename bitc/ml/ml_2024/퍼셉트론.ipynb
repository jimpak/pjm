```python
import numpy as np
```

```python
def AND_gate(x):
    w1=0.5
    w2=0.5
    b=-0.7
    result = x[0]*w1 + x[1]*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
def OR_gate(x):
    w1=0.6
    w2=0.6
    b=-0.5
    result = x[0]*w1 + x[1]*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
def NAND_gate(x):
    w1=-0.5
    w2=-0.5
    b=0.7
    result = x[0]*w1 + x[1]*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
input_data=np.array([[0,0],[0,1],[1,0],[1,1]])
result_and = []
for x in input_data:
    result_and.append(AND_gate(x))
result_and
```

```python
result_or = []
for x in input_data:
    result_or.append(OR_gate(x))
result_or
```

```python
result_nand = []
for x in input_data:
    result_nand.append(NAND_gate(x))
result_nand
```

```python
s1=[]
s2=[]
new_input_data=[]
final_output=[]

for i in range(len(input_data)):
    s1=NAND_gate(input_data[i])
    s2=OR_gate(input_data[i])
    new_input_data.append(s1)
    new_input_data.append(s2)
    result=AND_gate(new_input_data)
    final_output.append(result)
    new_input_data=[]

print(final_output)
```

```python
import torch
import torch.nn as nn
```

```python
X = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y = torch.FloatTensor([[0],[1],[1],[0]])

linear = nn.Linear(2,1,bias=True)
sigmoid = nn.Sigmoid()
model = nn.Sequential(linear, sigmoid)

criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)

    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()

    if step % 100 == 0:
        print(step, cost.item())
```

```python
prediction=model(X)
prediction
```

```python
model=nn.Sequential(
    nn.Linear(2,10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10,10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10,10, bias=True),
    nn.Sigmoid(),
    nn.Linear(10,1, bias=True),
    nn.Sigmoid()
)
```

```python
criterion=torch.nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(), lr=1)
```

```python
for epoch in range(10001):
    optimizer.zero_grad()
    y_hat=model(X)
    cost=criterion(y_hat, Y)
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(epoch, cost.item())
```
